```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from collections import Counter


# Load CSV file (change the filename as needed)
df = pd.read_csv("/content/Untitled form (Responses) - Form Responses 1.csv")

# Print the available columns to verify the correct column name
print(df.columns)

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from collections import Counter


# Load CSV file (change the filename as needed)
df = pd.read_csv("/content/Untitled form (Responses) - Form Responses 1.csv")

# Print the available columns to verify the correct column name
print(df.columns)

# Select the column to analyze (Update column name if needed - based on the printed columns and data)
# The correct column name is '1. Why do plastics stay longer in the environment?'
column_name =  '13.  Besides PET bottles and pure water sachets, what other plastic materials are commonly found the beaches and in other parts of water bodies?'
responses = df[column_name].dropna().astype(str).tolist()

# Convert text to numerical data (TF-IDF)
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(responses)

# Clustering similar responses
dbscan = DBSCAN(eps=0.5, min_samples=2, metric="cosine")
labels = dbscan.fit_predict(X)

# Organizing clusters
clusters = {}
for i, label in enumerate(labels):
    clusters.setdefault(label, []).append(responses[i])

# Count occurrences per cluster
cluster_counts = {k: len(v) for k, v in clusters.items()}

# Sort and display results
sorted_clusters = sorted(cluster_counts.items(), key=lambda x: x[1], reverse=True)

print("\nðŸ”¹ **Clustered Responses & Counts:**")
for cluster, count in sorted_clusters:
    if cluster != -1:  # Ignore unclustered responses
        print(f"\nCluster {cluster} ({count} responses):")
        print(clusters[cluster][:3], "...")  # Show sample responses

# Count ungrouped responses
noise_count = Counter(labels)[-1]
print(f"\nðŸ”¹ Unclustered Responses: {noise_count}")
# Convert text to numerical data (TF-IDF)
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(responses)

# Clustering similar responses
dbscan = DBSCAN(eps=0.5, min_samples=2, metric="cosine")
labels = dbscan.fit_predict(X)

# Organizing clusters
clusters = {}
for i, label in enumerate(labels):
    clusters.setdefault(label, []).append(responses[i])

# Count occurrences per cluster
cluster_counts = {k: len(v) for k, v in clusters.items()}

# Sort and display results
sorted_clusters = sorted(cluster_counts.items(), key=lambda x: x[1], reverse=True)

print("\nðŸ”¹ **Clustered Responses & Counts:**")
for cluster, count in sorted_clusters:
    if cluster != -1:  # Ignore unclustered responses
        print(f"\nCluster {cluster} ({count} responses):")
        print(clusters[cluster][:3], "...")  # Show sample responses

# Count ungrouped responses
noise_count = Counter(labels)[-1]
print(f"\nðŸ”¹ Unclustered Responses: {noise_count}")
```
